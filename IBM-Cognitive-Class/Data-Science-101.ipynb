```python
filename = "https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBMDeveloperSkillsNetwork-PY0101EN-SkillsNetwork/labs/Module%204/data/TopSellingAlbums.csv"

ibmDs = pd.read_csv(filename)
ibmDs.head(4)
```

```python
'''
  @brief: Getting the Unique Genres in the Dataset.
'''
for index, genre in enumerate(ibmDs['Genre'].unique().tolist()):
  print(f"Genre ({index + 1}) : {genre.title()}")
```

# **Learning NumPy Library**

```python
import numpy as np
```

```python
print(f"The Version of NumPy: {np.__version__}")
```

- **1-D Arrays**

```python
Arr = np.array([87, 67, 89, 89, 67])
ArrFloat = np.array([12.2, 67.4, 89.5, 45.7, 67.9])

# Accessing The Arrays!
print(f"The First Value: {Arr[0]}")
print(f"The Size (Arrays): {Arr.size}")
print(f"The Shape (Arrays): {Arr.shape}")
print(f"The Dimension (Arrays): {Arr.ndim}")


# Accesing Float Arrays!
print()
print(f"The Array 1: {Arr}")
print(f"The Array 2: {ArrFloat}")


# Performing Operations!-
print()
Sum = Arr + ArrFloat
print(f"The Sum of Arrays:{Sum}")
```

- Vector Calculation!

```python
# Following syntax: (x, y)
A = np.array([0, 1])
B = np.array([1, 0])

C = A + B
print(f"Vector Calculation: {C}")
```
