# **Learning NumPy Library**

```python
print(f"The Version of NumPy: {np.__version__}")
```

- **1-D Arrays**

```python
Arr = np.array([87, 67, 89, 89, 67])
ArrFloat = np.array([12.2, 67.4, 89.5, 45.7, 67.9])

# Accessing The Arrays!
print(f"The First Value: {Arr[0]}")
print(f"The Size (Arrays): {Arr.size}")
print(f"The Shape (Arrays): {Arr.shape}")
print(f"The Dimension (Arrays): {Arr.ndim}")


# Accesing Float Arrays!
print()
print(f"The Array 1: {Arr}")
print(f"The Array 2: {ArrFloat}")


# Performing Operations!-
print()
Sum = Arr + ArrFloat
print(f"The Sum of Arrays:{Sum}")


del(Arr, ArrFloat, Sum)
```

- Vector Calculation!

```python
# Following syntax: (x, y)
A = np.array([0, 1])
B = np.array([1, 0])

C = A + B
print(f"Vector Calculation: {C}")
del(A, B, C)
```

- Using **Linspace**
- This Method allows to get `num="n"` equally sepereated values from **high** and **low** value.

```python
lin = np.linspace(-4, 4, num=3)
print(f"The Array: {lin}")

del(lin)
```

```python
ln = np.linspace(0, 2*np.pi, num=100)
nl = np.sin(ln)
plt.plot(ln, nl)

del(ln, nl)
```

- **2-D Arrays**

```python
n1 = np.array([[1, 2], [3, 4], [4, 5]])
print(f"The Array: {n1}")
print(f"Dimension: {n1.ndim}")
print(f"Shape: {n1.shape}")

# Accessing The Elements in the 2-D Array
print(n1[0][1])
```

- **Slicing** 2-D Arrays!

```python
print(f"Printing Complete Array: \n{n1[0:][0:]}")
print(f"Printing Partial- Array: \n{n1[1:][0:]}")
print(f"Printing Last Part Array: \n{n1[:-1][:]}")

del(n1)
```

- **Operations** 2-D Arrays


```python
arr1 = np.array([[1, 3],  [2, 4]])
arr2 = np.array([[2, 3], [5, 1]])

arr3 = arr1 + arr2
arr4 = 2 * arr1
arr5 = np.dot(arr1, arr2)

# Adds All Elements of Arr1 & Arr2.
print(f"The Addition of Arr1 & Arr2 is \n{arr3}")
# Scalar Multiplication
print(f"The Multiplication of Arr1 with \"2\" \n{arr4}")
# Dot Product
print(f"The Dot Product of Arr1 & Arr2 is \n{arr5}")

del(arr1, arr2, arr3, arr4)
```
